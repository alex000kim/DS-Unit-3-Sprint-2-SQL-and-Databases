```python
import sqlite3
import pymongo
import pandas as pd
```

```python
import requests
import json

url = 'https://cat-fact.herokuapp.com/facts'
response = requests.request("GET", url)
```

```python
cat_fact_data = json.loads(response.text)['all']
```

```python
cat_fact_data[:5]
```

```python
client = pymongo.MongoClient()
```

```python
db = client.test_db
```

```python
cat_facts = db.cat_facts
```

```python
res = cat_facts.insert_many(cat_fact_data)
```

```python
len(res.inserted_ids) == len(cat_fact_data)
```

```python
list(cat_facts.find({'user.name.first': 'Alex',
                     'user.name.last': 'Wohlbruck'}))
```

```python
facts_dict = {
    'factID': [],
    'userID': [],
    'text': [],
    'upvotes': []
}

user_dict = {
    'userID': [],
    'firstName': [],
    'lastName': []
}
```

```python
for ii in cat_fact_data:
    facts_dict['factID'].append(ii['_id'])
    facts_dict['text'].append(ii['text'])
    facts_dict['upvotes'].append(ii['upvotes'])
    if 'user' in ii:
        user_id = ii['user']['_id']
        facts_dict['userID'].append(user_id)
        if user_id not in user_dict['userID']:
            user_dict['userID'].append(ii['user']['_id'])
            user_dict['firstName'].append(ii['user']['name']['first'])
            user_dict['lastName'].append(ii['user']['name']['last'])
    else:
        facts_dict['userID'].append(None)
```

```python
df_facts = pd.DataFrame(facts_dict)
df_users = pd.DataFrame(user_dict)
```
